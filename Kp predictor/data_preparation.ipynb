## **OMNI DATA**

```python
import pandas as pd

# Read data from file
file_path = "Data Files/OMNIWEB_data(1967 - 2023).txt"

with open(file_path, "r") as f:
    lines = f.readlines()

# Split the lines and extract the data
data = [line.strip().split() for line in lines]

# Create DataFrame
data_list = [
    "YEAR",
    "DOY",
    "HR",
    "Scalar B, nT",
    "Vector B Magnitude, nT",
    "Lat. Angle of B (GSE)",
    "Long. Angle of B (GSE)",
    "BX, nT (GSE, GSM)",
    "BY, nT (GSM)",
    "BZ, nT (GSM)",
    "SW Proton Density, N/cm^3",
    "SW Plasma Speed, km/s",
    "Kp index",
    "R (Sunspot No.)",
    "ap_index, nT",
    "f10.7_index"
]

omni_data = pd.DataFrame(data, columns=data_list)

# Convert columns to appropriate data types
omni_data = omni_data.astype({
    "YEAR": int,
    "DOY": int,
    "HR": int,
    "Scalar B, nT": float,
    "Vector B Magnitude, nT": float,
    "Lat. Angle of B (GSE)": float,
    "Long. Angle of B (GSE)": float,
    "BX, nT (GSE, GSM)": float,
    "BY, nT (GSM)": float,
    "BZ, nT (GSM)": float,
    "SW Proton Density, N/cm^3": float,
    "SW Plasma Speed, km/s": float,
    "Kp index": int,
    "R (Sunspot No.)": int,
    "ap_index, nT": int,
    "f10.7_index": float
})

# Print the DataFrame
# print(wind_df)

# Export the DataFrame to CSV
omni_data.to_csv("omni_data.csv", index=False)
```

## **Kp DATA**

```python
import pandas as pd

# Read the data from the txt file
file_path = "Data files/raw_data/kp_data(2).txt"

data = []
with open(file_path, "r") as file:
    for line in file:
        row = line.strip().split()
        # Convert numerical elements to float and keep the rest as strings
        row = [float(item) if item.replace(".", "", 1).isdigit() else item for item in row]
        data.append(row)

# Define column names based on the provided data
column_names = [
    "YYY",
    "MM",
    "DD",
    "hours",
    "hours_mins",
    "days",
    "days_m",
    "Kp",
    "ap",
    "D"
]


# Create DataFrame from the data list
kp_df = pd.DataFrame(data, columns=column_names)

# Display the DataFrame
print(kp_df)
```

```python
kp_df = kp_df.astype({
   "YYY": int,
    "MM": int,
    "DD":int,
    "hours":float,
    "hours_mins":float,
    "days":float,
    "days_m":float,
    "Kp": float,
    "ap": int,
    "D":int
})

# Print the DataFrame
kp_df
```
